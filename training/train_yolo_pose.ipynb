```python
from ultralytics import YOLO
import torch
```

```python
print(torch.cuda.is_available())
print(torch.cuda.device)
```

```python
# 1. Load mô hình pretrained (bản Nano cho nhẹ và nhanh)
model = YOLO('model_loader/yolov8n-pose.pt')

# 2. Huấn luyện
results = model.train(
    data='dataset/26_05_18_Truck_Pose_01_dataset/data.yaml', 
    epochs=150,      # Số vòng lặp, PoC nên để 100-150
    imgsz=640,       # Kích thước ảnh đầu vào
    device=0,        # Sử dụng GPU (nếu có)
    project='26-05-18-truck-pose', # Tên thư mục lưu kết quả
    name='poc_v1'    # Tên đợt training này
)
```
